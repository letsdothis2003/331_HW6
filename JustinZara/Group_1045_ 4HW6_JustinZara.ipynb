<span style="color:#808080;">--proposition 1: select the names and ID's of employees hired after </span> <span style="color: rgb(128, 128, 128); text-decoration-line: underline; text-decoration-color: rgb(160, 160, 160); text-decoration-style: wavy;">january</span> <span style="color:#808080;">2, 2009, and also with the job senior design engineer</span>

```sql
USE AdventureWorks2022;


SELECT e.BusinessEntityID,CONCAT(p.FirstName, ' ',p.LastName), e.HireDate, e.JobTitle
FROM HumanResources.Employee as e
INNER JOIN Person.Person p 
ON e.BusinessEntityID = p.BusinessEntityID
WHERE e.HireDate > '2009-01-02'
INTERSECT
SELECT e.BusinessEntityID,CONCAT(p.FirstName, ' ',p.LastName), e.HireDate, e.JobTitle 
FROM HumanResources.Employee as e
INNER JOIN Person.Person p 
ON e.BusinessEntityID = p.BusinessEntityID
WHERE e.JobTitle = 'Senior Design Engineer';
```

<span style="color:#808080;">* It intersects rows where the person's job title is senior design engineer, with rows that have a hire date hire date after January 2 2009</span>

<span style="color:#808080;">* *This query can be used by businesses to assess how long employees have been employed.*/</span>

<span style="color:#808080;">--proposition 2: select the names and ID's of employees who has greater than 90 vacation hours, or employees that have greater than 50 sick hours</span>

```sql
SELECT e.BusinessEntityID , e.VacationHours, e.SickLeaveHours
FROM HumanResources.Employee as e
WHERE e.VacationHours > 90
UNION
SELECT e.BusinessEntityID, e.VacationHours, e.SickLeaveHours
FROM HumanResources.Employee as e
WHERE e.SickLeaveHours  > 50;
```

<span style="color:#808080;">--Proposition 3: select the product ID's that have a list price greater than 3500 except products that have a weight less than 14.5</span>

```sql
SELECT p.ProductID 
FROM Production.Product p
WHERE p.ListPrice > 3500
EXCEPT
SELECT p.ProductID 
FROM Production.Product p
WHERE p.Weight < 14.5;
```

<span style="color:#808080;">--Proposition 4: Select the ID of special offers that start in the year 2012 except offers in April</span>

```sql
SELECT so.SpecialOfferID, so.StartDate, so.Description 
FROM Sales.SpecialOffer so
WHERE YEAR(so.StartDate) = '2012'
EXCEPT 
SELECT so.SpecialOfferID, so.StartDate, so.Description 
FROM Sales.SpecialOffer so
WHERE MONTH(so.StartDate) = '04';
```

<span style="color:#808080;">--Proposition 5: Get the list of ID's of employee departments that start at 2010, except for those that have a shift ID of 1.</span>

```sql
SELECT edh.BusinessEntityID , edh.StartDate, edh.ShiftID 
FROM HumanResources.EmployeeDepartmentHistory edh
WHERE YEAR(edh.StartDate) = 2010
EXCEPT
SELECT edh.BusinessEntityID , edh.StartDate, edh.ShiftID 
FROM HumanResources.EmployeeDepartmentHistory edh
WHERE edh.ShiftID = 1;
```

<span style="color:#808080;">/*This query outputs the ID, start date, and shift id of employees who started in 2010, and excludes employees who has a shift ID of 1.</span>

<span style="color:#808080;">* This can be used by businesses to keep track of the employees or departments that are assigned to specific shifts.*/</span>

<span style="color:#808080;">--Proposition 10 list out products that are red, or have a style of U.</span>

```sql
SELECT p.ProductID, p.Color, p.[Style] 
FROM  Production.Product p
WHERE p.Color = 'Red'
UNION
SELECT p.ProductID, p.Color, p.[Style] 
FROM  Production.Product p
WHERE p.Style = 'U'
```
